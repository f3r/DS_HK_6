## Run Regression

```python
print xlist
```

```python
xmatrix = df[['revert1', 'revert2', 'revert3', 'revert4', 'revert5', 'revert6', 'revert7', 'revert8', 'revert9', 'mo1', 'mo2', 'mo3', 'mo4', 'mo5', 'mo6', 'mo7', 'mo8', 'mo9', 'mo10', 'mo11']]
```

```python
from statsmodels.formula.api import ols

results = ols('return0 ~ xmatrix', df).fit()
results.summary()
```

The overall R square is not high. But several factors have low P value. Let's explain if we only use variables where P < 0.10

```python
#Reducing the number of x variables & rerun regression

xmatrix2 = df[['revert1', 'revert2', 'revert3', 'revert4', 'revert9', 'mo6']]
results2 = ols('return0 ~ xmatrix2', df).fit()
results2.summary()
```

```python
results2.params
```

```python
#Make a multi variable regression statement

df['predict'] = results2.params[0] + results2.params[1] * df[xmatrix2.columns[0]] \
    + results2.params[2] * df[xmatrix2.columns[1]] + results2.params[3] * df[xmatrix2.columns[2]] \
    + results2.params[4] * df[xmatrix2.columns[3]] + results2.params[5] * df[xmatrix2.columns[4]]
```

```python
#Plot the predicted return vs actual return (y)
sns.lmplot('predict', 'return0', df)
```

```python
#Looks like some correlation. Create strategy where you Long when predicted return > 0, and vice versa
import numpy as np

df['pnl'] = np.where(df['predict']> 0 , df['return0'], np.where(df['predict'] < 0 , -df['return0'], 0 ))
```

```python
#heavier positive returns

ggplot(df, aes(x = 'pnl')) + geom_density()
```

```python
#Some stats on the daily P&L
df['pnl'].describe()
```

```python
#Looking at the cumulative PnL

df['cum_pnl'] = df.pnl.cumsum()
plt.plot(df['cum_pnl'])
```

##Out Sample Data

```python
df2 = pd.read_csv("DS_stock_dataset_outsample.csv")

df2['predict'] = results2.params[0] + results2.params[1] * df2[xmatrix2.columns[0]] \
    + results2.params[2] * df2[xmatrix2.columns[1]] + results2.params[3] * df2[xmatrix2.columns[2]] \
    + results2.params[4] * df2[xmatrix2.columns[3]] + results2.params[5] * df2[xmatrix2.columns[4]]
    
df2['pnl'] = np.where(df2['predict']> 0 , df2['return0'], np.where(df2['predict'] < 0 , -df2['return0'], 0 ))

df2['cum_pnl'] = df2.pnl.cumsum()
```

```python
plt.plot(df2['cum_pnl'])
```
